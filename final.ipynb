```python
artist_danceability = dict(final.groupby('artist')['danceability'].mean())
```

```python
artist_danceability_sorted = sorted(artist_danceability.items(), key=lambda kv: kv[1])
```

```python
plt.subplots(figsize=(10,5))
dance_plot = sns.barplot(x=[artist[0] for artist in artist_danceability_sorted[-20:]], y=[artist[1] for artist in artist_danceability_sorted[-20:]])
dance_plot.tick_params(labelrotation=90)
dance_plot.set_title('Most Danceabe Artists')
plt.ylabel('Spotify - Danceability', fontsize=12)
plt.xlabel('Artist', fontsize=12)
```

```python
hit_count = final.groupby('artist')['is_hit']
hit_count = dict(hit_count.sum())
import operator
sortedhits= (sorted(hit_count.items(), key=operator.itemgetter(1),reverse=True))
sortedhits= sortedhits[0:12]
sortedhits = dict(sortedhits)
plt.figure(figsize=(10,5))
sns.barplot(list(sortedhits.keys()), list(sortedhits.values()), alpha=0.8)
plt.title('Top 12 Artists by HitCount')
plt.ylabel('Hits', fontsize=12)
plt.xlabel('Artist', fontsize=7)
plt.tick_params(labelrotation=30)
plt.show()
```
